```python
import pandas as pd
from tqdm import tqdm
import nltk
from nltk.tokenize import RegexpTokenizer
from konlpy.tag import Mecab
mecab = Mecab(dicpath=r"C:/mecab/mecab-ko-dic")
```

```python
stop_word = "전 난 일 걸 뭐 줄 만 건 작업 분 위 개 끝 송 잼 이거 부 동 번 중 듯 차 때 게 내 말 나 수 거 점 것 등 측 의 급 후 간 단 시 곳 에"
stop_word = stop_word.split(' ')
```

```python
df = pd.read_excel('./source/working_standard.xlsx')
df
```

```python
df_act = df['작업행동']
```

```python
targets = []
results = []
```

```python
for sentence in tqdm(df_act):
    sentence = str(sentence)
    result = []
#     tokenizer = RegexpTokenizer("[\w]+")
    tokenizer = RegexpTokenizer("^[가-힣]")
    tokens = tokenizer.tokenize(sentence)
    for token in tokens:
        if token not in stop_word:
            result.append(token)
    targets.append(sentence)
    results.append(result)
```
