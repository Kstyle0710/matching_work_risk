### 작업행동과 위험벡터 직접 맵핑

```python
works2 = copy.deepcopy(works)
works2['형태소']
```

```python
vectors
```

```python
for n1 in range(len(works2["형태소"])):
    k = works2['형태소'].iloc[n1]
    k = k.split(",")
#     print(k)
#     print(type(k))
    point = 0
    for i in k:
        i = i.strip()
        print(i)
        if i in list(vectors['키워드']):
            point += float(vectors.loc[vectors['키워드'] == i, '위험지수'])
        else:
            point += 0
    works2.at[n1, "위험점수"] = point 
```
